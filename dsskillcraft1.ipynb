```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (assuming it's a CSV file)
# Replace 'your_travel_time_data.csv' with the actual file path
df = pd.read_csv('/content/Final_Majestic_to_AIM_jan-2016tomarch-2018.csv')

# Set a colorful palette for visualizations
sns.set_palette("husl")

# List of features to visualize
travel_time_features = [
    'Daily Mean Travel Time (Seconds)',
    'AM Mean Travel Time (Seconds)',
    'PM Mean Travel Time (Seconds)',
    'Midday Mean Travel Time (Seconds)',
    'Evening Mean Travel Time (Seconds)',
    'Early Morning Mean Travel Time (Seconds)'
]

# Plot histograms for each travel time feature
for feature in travel_time_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.show()

# Optionally, you can create bar plots for specific travel times or daily averages
plt.figure(figsize=(10, 6))
sns.barplot(x='Date', y='Daily Mean Travel Time (Seconds)', data=df, palette="coolwarm")
plt.title('Daily Mean Travel Time Over Dates', fontsize=15)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Mean Travel Time (Seconds)', fontsize=12)
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.show()
```
